```python
combined.info
```

```python
##splitting dependent and independent variable

output=combined['Activity'].copy()
input=combined.drop(['Activity','subject'],axis=1).copy()
```

#####Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
slc=StandardScaler()
inp=slc.fit_transform(input)
```

#####Doing dimensionalit reduction ==>PCA


*   we have 563 variables (columns) in our data. Some of them might not contribute to give out the best result from the model.
*   Applying PCA to reduce the dimensions from 563 to 65.



```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=65)
```

```python
feature=pca.fit_transform(inp)
feature.shape
```

```python
feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,output,test_size=0.2,random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
output.unique()
```

#####USing KNN model to predict the activity

> Indented block



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  classification_report,accuracy_score,confusion_matrix
```

```python
knn=KNeighborsClassifier(5)   #5 neigbours
knn.fit(x_train,y_train)
```

```python
predict=knn.predict(x_test)
predict[:5]
```

```python
y_test[:5]
```

```python
(y_test==predict).sum()/2060
```

```python
print(accuracy_score(y_test,predict))
```

```python
print(confusion_matrix(y_test, predict))
```

```python
classification_report(y_test,predict)
```
